```
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = load_iris()
```

```
df = pd.DataFrame(data=data['data'], columns=data['feature_names'])
df['target'] = pd.Categorical.from_codes(data.target, data.target_names)
```

```
df.head()
```

Number of features and their types.

```
y = df['target']
x = df.drop(['target'],axis=1)
```

```
x.dtypes
```

```
y.value_counts()
```

Displaying summary statistics for each column

```
x.describe()
```

Data Visualisation

```
plt.hist(x['sepal length (cm)'],color='pink')
plt.title('Variaions in sepal length')
plt.xlabel('Sepal length')
plt.ylabel('Frequency')
plt.show()
```

```
plt.hist(x['sepal width (cm)'])
plt.title('Variaions in sepal width')
plt.xlabel('Sepal width')
plt.ylabel('Frequency')
plt.show()
```

```
plt.hist(x['petal length (cm)'],color="brown")
plt.title('Variations in petal length')
plt.xlabel('Petal length')
plt.ylabel('Frequency')
plt.show()
```

```
plt.hist(x['petal width (cm)'],color="green")
plt.title('Variations in petal width')
plt.xlabel('Petal width')
plt.ylabel('Frequency')
plt.show()
```

BoxPlot for each feature

```
df.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```
x.boxplot()
```

```
sns.FacetGrid(df,hue="target",size=3).map(sns.distplot,"petal length (cm)").add_legend()
sns.FacetGrid(df,hue="target",size=3).map(sns.distplot,"petal width (cm)").add_legend()
sns.FacetGrid(df,hue="target",size=3).map(sns.distplot,"sepal length (cm)").add_legend()
sns.FacetGrid(df,hue="target",size=3).map(sns.distplot,"sepal width (cm)").add_legend()
plt.show()
```

```
sns.violinplot(x="target",y="petal length (cm)",data=df)
plt.show()
```

Pair Plots are a really simple (one-line-of-code simple!) way to visualize relationships between each variable. It produces a matrix of relationships between each variable in your data for an instant examination of our data.

pair plot gives scatter plot of different features.

pair plot for iris data set.


```
sns.set_style("whitegrid")
sns.pairplot(df,hue="target",size=5);
plt.show()
```
